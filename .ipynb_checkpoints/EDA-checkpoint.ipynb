```python
# check post_message: duplicate
index = data[(data.duplicated(subset=['post_message'], keep='last') == True)].index
nums = []
for i in index:
    data_duplicate = data[(data.post_message == data.post_message[i])]
    nums.append(len(data_duplicate))
    print(f'Data at index: {i}')
    print(data_duplicate)
    print('-'*170)
```

```python
for (i, num) in zip(index, nums):
    print('*'*170)
    print(f'post_message at index {i}, Number of duplicate {num}:')
    print(data.post_message[i])
```
