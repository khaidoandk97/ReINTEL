```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from utils import *
from model import *
```

```python
num_data = pd.read_csv('num_data_del.csv', encoding='utf8')
num_test = pd.read_csv('num_test.csv', encoding='utf8')
```

```python
cols = ['num_like_post', 'num_comment_post', 'num_share_post']
X, y = num_data[cols].values, num_data.label.values
X_test, y_test = num_test[cols].values, num_test.label.values

print(X.shape, y.shape)
print(X_test.shape, y_test.shape)
```

```python
options = [0,1,2,3,4,5]
names = ['KNN','Gaussian', 'SVM', 'Randomforest', 'Logistc', 'Stack']

for (option, name) in zip(options, names):
    print(f'{name} model:')
    model = models(X, y, option=option)
    _ =  predict(model=model, X=X_test, y=y_test)
    print('*'*150)
```
