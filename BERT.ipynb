#3. OVER SAMPLING

```
# import libraries
# check version number of sampling library
import imblearn
print(imblearn.__version__)

from collections import Counter
from imblearn.over_sampling import RandomOverSampler
import pandas as pd
import numpy as np

# take data
data = pd.read_csv('train.csv', encoding='utf-8')
test = pd.read_csv('test.csv', encoding='utf-8')

train_sents = data.post_message.values
train_labels = data.label.values

test_text = test.post_message_preproced.values
test_labels = test.label.values

# define oversampling startegy 
oversample = RandomOverSampler(sampling_strategy=0.3)

# fit and apply the transform
train_sents_over, train_labels_over = oversample.fit_resample(train_sents.reshape(-1,1), train_labels.reshape(-1,1))
train_sents_over, train_labels_over = train_sents_over.reshape(-1), train_labels_over.reshape(-1)
# summarize class distribution
print(Counter(train_labels_over))
```

```
# Train model with Bert_original_bpe_tfidf_overSampling
path="./models/Bert_original_bpe_tfidf_overSampling.pt"

BERT, vectorizer = bert_train(train_sents_over, train_labels_over, test_text, test_labels, device = 'cuda', epochs = 10, path=path, w2v='bpe_tfidf')

# predict for test set 
test_labels_pred = bert_predict(test_text, BERT, vectorizer, w2v='bpe_tfidf')
print('*'*100)
report(test_labels_pred, test_labels, name='Test')
```
